```python
import cv2 as cv2
import numpy as np
from PIL import Image
import os
from pathlib import Path
import argparse
import random
import numpy as np
import matplotlib.cm as cm
import torch
from skimage.transform import resize
import glob
import openslide
import matplotlib.pyplot as plt
import xmltodict
import pandas as pd
from PIL import ImageFile
from skimage.transform import resize
from PIL import PngImagePlugin
```

```python
## 老师方法
LARGE_ENOUGH_NUMBER = 100
PngImagePlugin.MAX_TEXT_CHUNK = LARGE_ENOUGH_NUMBER * (1024**2)

directory = ['5X', '10X', '40X']

def get_none_zero(black_arr):
    nonzeros = black_arr.nonzero()
    starting_y = nonzeros[0].min()
    ending_y = nonzeros[0].max()
    starting_x = nonzeros[1].min()
    ending_x = nonzeros[1].max()
    return starting_x, starting_y, ending_x, ending_y

def scan_nonblack(simg, px_start, py_start, px_end, py_end):
    offset_x = 0
    offset_y = 0
    line_x = py_end - py_start
    line_y = px_end - px_start

    val = simg.read_region((px_start + offset_x, py_start), 0, (1, 1))
    arr = np.array(val)[:, :, 0].sum()
    while arr == 0:
        val = simg.read_region((px_start + offset_x, py_start), 0, (1, line_x))
        arr = np.array(val)[:, :, 0].sum()
        offset_x = offset_x + 1

    val = simg.read_region((px_start, py_start + offset_y), 0, (1, 1))
    arr = np.array(val)[:, :, 0].sum()
    while arr == 0:
        val = simg.read_region((px_start, py_start + offset_y), 0, (line_y, 1))
        arr = np.array(val)[:, :, 0].sum()
        offset_y = offset_y + 1

    x = px_start + offset_x - 1
    y = py_start + offset_y - 1
    return x, y

def preprocess_all():
    for dirpath, _, files in os.walk('svs'):
        if len(files) > 0:
            for dirname in directory:
                path = os.path.join(dirpath, dirname)
                os.makedirs(path, exist_ok=True)

    for dirpath, dirnames, files in os.walk('svs'):
        print(f'pre-processing directory: {dirpath}')
        for file_name in files:
            print(file_name)
            img = openslide.open_slide(dirpath + '/' + file_name)
            x, y = scan_nonblack(img, 0, 0, img.dimensions[0], img.dimensions[1])
            start = img.dimensions - np.array((x, y))

            filename_40X = dirpath + '/' + '40X/40X_' + file_name.replace('.svs', '.png')
            img.read_region(start, 0, img.dimensions).save(filename_40X)

            img_40X = plt.imread(filename_40X)

            img_10X = resize(img_40X, (int(img_40X.shape[0] / 4), int(img_40X.shape[1] / 4)))
            filename_10X = dirpath + '/' + '10X/10X_' + file_name.replace('.svs', '.png')
            plt.imsave(filename_10X, img_10X)

            img_5X = resize(img_40X, (int(img_40X.shape[0] / 8), int(img_40X.shape[1] / 8)))
            filename_5X = dirpath + '/' + '5X/5X_' + file_name.replace('.svs', '.png')
            plt.imsave(filename_5X, img_5X)

def preprocess_one(dirpath, filename):
    # 打开SVS图像
    img = openslide.open_slide(dirpath + '/' + filename)
    
    # 获取图像区域的起始位置
    x, y = scan_nonblack(img, 0, 0, img.dimensions[0], img.dimensions[1])
    start = img.dimensions - np.array((x, y))

    # 创建保存目录
    folder_10X = dirpath + '/' + '10X'
    folder_5X = dirpath + '/' + '5X'
    if not os.path.exists(folder_10X):
        os.makedirs(folder_10X)
    if not os.path.exists(folder_5X):
        os.makedirs(folder_5X)

    # 保存10X图像为JPEG
    filename_10X = folder_10X + '/' + '10X_' + filename.replace('.svs', '.jpg')
    print('saving 10X jpeg...')
    
    # 读取图像区域，调整为10X尺寸，并保存为JPEG格式
    img_region = img.read_region(start, 0, img.dimensions)
    img_10X = resize(np.array(img_region), (int(img_region.size[1] / 4), int(img_region.size[0] / 4)))
    img_10X = Image.fromarray(img_10X)
    
    # 转换为 RGB 格式（去掉透明度通道）
    img_10X = img_10X.convert("RGB")
    img_10X.save(filename_10X, format='JPEG')

    # 保存5X图像为JPEG
    filename_5X = folder_5X + '/' + '5X_' + filename.replace('.svs', '.jpg')
    print('saving 5X jpeg...')
    
    # 进一步缩小为5X尺寸，并保存为JPEG格式
    img_5X = resize(np.array(img_10X), (int(img_10X.size[1] / 2), int(img_10X.size[0] / 2)))
    img_5X = Image.fromarray(img_5X)
    
    # 转换为 RGB 格式（去掉透明度通道）
    img_5X = img_5X.convert("RGB")
    img_5X.save(filename_5X, format='JPEG')


def test(dirpath, filename):
    img = openslide.open_slide(dirpath + '/' + filename)
    print(img.properties)

if __name__ == '__main__':
    dirpath = '/Users/zz/Desktop/reser/ruiming/pathology/public_data'
    filename = 'TX1_HE1.svs'

    preprocess_one(dirpath, filename)
```

```python
# 魔改但没法resize

def preprocess_one(dirpath, filename):
    """ Convert SVS to PNG at 40X, 10X, and 5X magnifications efficiently. """

    # 打开SVS文件
    slide = openslide.OpenSlide(os.path.join(dirpath, filename))
    width, height = slide.dimensions
    print(f"📌 Processing: {os.path.basename(filename)}, size: {width}x{height}")

    # 使用 OpenSlide 提供的内建缩放来有效加载图像
    downsample_factor = 4  # 加载 1/4 分辨率以减少内存占用
    img_40X = np.array(slide.read_region((0, 0), 0, (width // downsample_factor, height // downsample_factor)).convert("RGB"))

    # 保存 40X 图像为 PNG
    filename_40X = os.path.join(dirpath, "40X", os.path.basename(filename).replace(".svs", "_40X.png"))
    os.makedirs(os.path.dirname(filename_40X), exist_ok=True)
    img_40X_pil = Image.fromarray(img_40X)
    img_40X_pil.save(filename_40X, format="PNG")

    # 生成并保存 10X（缩小为 1/4）
    img_10X = resize(img_40X, (img_40X.shape[0] // 4, img_40X.shape[1] // 4), anti_aliasing=True)
    filename_10X = os.path.join(dirpath, "10X", os.path.basename(filename).replace(".svs", "_10X.png"))
    os.makedirs(os.path.dirname(filename_10X), exist_ok=True)
    img_10X_pil = Image.fromarray((img_10X * 255).astype(np.uint8))  # 转换为 uint8 类型
    img_10X_pil.save(filename_10X, format="PNG")

    # 生成并保存 5X（缩小为 1/8）
    img_5X = resize(img_10X, (img_10X.shape[0] // 2, img_10X.shape[1] // 2), anti_aliasing=True)
    filename_5X = os.path.join(dirpath, "5X", os.path.basename(filename).replace(".svs", "_5X.png"))
    os.makedirs(os.path.dirname(filename_5X), exist_ok=True)
    img_5X_pil = Image.fromarray((img_5X * 255).astype(np.uint8))  # 转换为 uint8 类型
    img_5X_pil.save(filename_5X, format="PNG")

    print(f"✅ Conversion complete: {filename_40X}, {filename_10X}, {filename_5X}")

    # 释放内存
    del img_40X, img_10X, img_5X
    gc.collect()

if __name__ == '__main__':
    dirpath = '/Users/zz/Desktop/reser/ruiming/pathology/public_data'  # 请替换为你的文件夹路径
    filename = 'TX1_HE1.svs'  # 请替换为你的文件名

    preprocess_one(dirpath, filename)
```

```python
def ToPNG(case_name, IMG, output_folder):
    """
    Converts a large SVS file into a single PNG, downsampled for efficiency.

    Parameters:
        - case_name (str): Name of the case (used in filenames)
        - IMG (str): Path to the SVS file
        - output_folder (str): Where to save the PNGs
    """
    output_dir = output_folder
    os.makedirs(output_dir, exist_ok=True)

    X20_folder = os.path.join(output_folder, '20X')
    img_20X_path = os.path.join(X20_folder, f"image_{case_name}.png")

    if os.path.exists(img_20X_path):
        return

    try:
        # Open the whole-slide image using OpenSlide
        svs = openslide.OpenSlide(IMG)
    except:
        print('Error loading SVS file')
        return

    # Get the original dimensions of the SVS file
    x_length, y_length = svs.dimensions

    # Define output folder for the magnification level (20X)
    X20_folder = os.path.join(output_folder, case_name)
    os.makedirs(X20_folder, exist_ok=True)

    # Get the magnification level from SVS properties
    mag = svs.properties['aperio.AppMag']

    if mag == '40':
        lv = 1
        down = 2
    elif mag == '20':
        lv = 0
        down = 1

    # Now, extract a region from the image and save it
    # Here we choose the full image at the appropriate magnification level.
    img_5X = np.array(svs.read_region((0, 0), lv, (int(x_length / down), int(y_length / down))))
    img_5X_path = os.path.join(X20_folder, f"{case_name}.png")
    
    # Save the extracted region as PNG
    cv2.imwrite(img_5X_path, cv2.cvtColor(img_5X, cv2.COLOR_RGB2BGR))

    # Free memory
    del img_5X
    svs.close()

if __name__ == "__main__":
    svs_dir = '/Users/zz/Desktop/reser/ruiming/pathology/public_data'
    output_dir = '/Users/zz/Desktop/reser/ruiming/pathology/public_data/png'

    svs_list = glob.glob(os.path.join(svs_dir, '*.svs'))

    for svs in svs_list:
        name = os.path.basename(svs)
        if not name.endswith('.svs'):
            continue

        case_name = name.split('.svs')[0]
        print(f"Processing: {case_name}")

        ToPNG(case_name, svs, output_dir)
```

```python
import os
import glob
import numpy as np
import openslide
import cv2

def ToPNG(case_name, IMG, output_folder):
    """
    Converts a large SVS file into a single PNG, downsampled for efficiency.

    Parameters:
        - case_name (str): Name of the case (used in filenames)
        - IMG (str): Path to the SVS file
        - output_folder (str): Where to save the PNGs
    """
    output_dir = output_folder
    os.makedirs(output_dir, exist_ok=True)

    X20_folder = os.path.join(output_folder, '20X')
    img_20X_path = os.path.join(X20_folder, f"image_{case_name}.png")

    if os.path.exists(img_20X_path):
        return

    try:
        # Open the whole-slide image using OpenSlide
        svs = openslide.OpenSlide(IMG)
    except:
        print('Error loading SVS file')
        return

    # Get the original dimensions of the SVS file
    x_length, y_length = svs.dimensions

    # Define output folder for the magnification level (20X)
    X20_folder = os.path.join(output_folder, case_name)
    os.makedirs(X20_folder, exist_ok=True)

    # Get the magnification level from SVS properties
    mag = svs.properties['aperio.AppMag']

    if mag == '40':
        lv = 1
        down = 2
    elif mag == '20':
        lv = 0
        down = 1

    # Here we process the image based on the magnification level
    img_5X_sample1 = np.array(svs.read_region((0, 0), lv, (int(x_length / down), int(y_length / down))))
    img_5X_path = os.path.join(X20_folder, f"{case_name}.png")
    
    # Save the extracted region as PNG
    cv2.imwrite(img_5X_path, cv2.cvtColor(img_5X_sample1, cv2.COLOR_RGB2BGR))

    # Free memory
    del img_5X_sample1
    svs.close()

if __name__ == "__main__":
    svs_dir = '/Users/zz/Desktop/reser/ruiming/pathology/public_data'
    output_dir = '/Users/zz/Desktop/reser/ruiming/pathology/public_data/png'

    svs_list = glob.glob(os.path.join(svs_dir, '*.svs'))

    for svs in svs_list:
        name = os.path.basename(svs)
        if not name.endswith('.svs'):
            continue

        case_name = name.split('.svs')[0]
        print(f"Processing: {case_name}")

        ToPNG(case_name, svs, output_dir)
```
